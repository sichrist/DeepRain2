```python
from PIL import Image
import numpy as np
import pandas as pd
import os
CSVFILE = "./.listOfFiles.csv"
WRKDIR = "./Data"
TRAINSETFOLDER=os.path.join(WRKDIR,"train")
VALSETFOLDER=os.path.join(WRKDIR,"val")

DatasetFolder = "./Data/train/256x320_256x320"
PathToData = os.path.join(DatasetFolder,"MonthPNGData")
```

```python
listOfFiles = list(pd.read_csv(os.path.join(DatasetFolder,".listOfFiles_train_.csv"))["colummn"])
```

```python
def sortOut(listOfFiles):
    y = []
    for i,path in enumerate(listOfFiles):
        img = np.array(Image.open(path))
        if img.max() > 0:
            y.append(i)
    return y
```

```python
y = sortOut(listOfFiles)
```
