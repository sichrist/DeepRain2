```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import pandas as pd
modelPath = "/home/simon/gitprojects/DeepRain2/Networks/model_data"
```

```python
def gethistory(folder,name):
    historypath = os.path.join(modelPath,folder)
    historypath = os.path.join(historypath,name)
    file = open(historypath,'r')
    json_str = file.read()
    history = json.loads(json_str)
    return history

def plotLoss(history,name):
    fig, axes = plt.subplots(1, figsize=(15,15),dpi=100, sharex=False, sharey=False)
    # Plot training & validation loss values
    axes.plot(history['loss'])
    axes.plot(history['val_loss'])
    axes.set_title(name)
    axes.set_ylabel('Loss')
    axes.set_xlabel('Epoch')
    axes.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```python
modelname = "LSTM_Meets_Unet_Upconv_mse272x224x5"
foldername = "LSTM_Meets_Unet_Upconv_mse"
history = gethistory(foldername,modelname+"history.json")


plotLoss(history,"LSTM_Meets_Unet_Upconv_mse272x224x5")

confpath = os.path.join(modelPath,foldername+"/validation/confusion.csv")
confmatrix = pd.read_csv(confpath)
print(confmatrix)

confl = list(confmatrix)
```
